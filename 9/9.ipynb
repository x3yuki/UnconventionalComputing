```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random as rd
from scipy.integrate import odeint
import time
```

```python
with open("log/log-1.txt", "r") as f:
    data = f.read().split("\n")
data = [float(d) for d in data if d != '']

plt.title("Crowding")
plt.plot(np.arange(0, len(data), 1), data, label='1', color='red')
plt.grid(which='major',color='gray',linestyle='-')
plt.ylim(-5, 0)
```
